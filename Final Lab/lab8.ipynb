**Data Preparation**

```python
# Drop columns not used for data because they're not helpful or too many null values
cols_to_drop = ['Standard deviation/Mean of ladder by country-year', 'GINI index (World Bank estimate)', 'GINI index (World Bank estimate), average 2000-15', 'gini of household income reported in Gallup, by wp5-year', 'Democratic Quality', 'Delivery Quality']
df = df.drop(cols_to_drop, axis=1)

# Drop columns before 2014
df = df[df['year'] >= 2014]

# Transform negative features to ensure 'comprehensive_wellbeing' is accurate
df['Perceptions of Corruption (Inverse)'] = 1 - df['Perceptions of corruption']
df['Negative Affect (Inverse)'] = 1 - df['Negative affect']
df['STD Dev of Ladder (Inverse)'] = 1 / df['Standard deviation of ladder by country-year']
old_cols_to_drop = ['Perceptions of corruption', 'Negative affect', 'Standard deviation of ladder by country-year']
df = df.drop(old_cols_to_drop, axis=1)

# Dropping null values for small percentages
columns_to_drop_nulls = [
    "Log GDP per capita",
    "Social support",
    "Healthy life expectancy at birth",
    "Freedom to make life choices",
    "Generosity",
    "Positive affect",
    "Confidence in national government",
    "Perceptions of Corruption (Inverse)",
    "Negative Affect (Inverse)"
]
df = df.dropna(subset=columns_to_drop_nulls)
```

```python
features = [
    "Life Ladder",
    "Log GDP per capita",
    "Social support",
    "Healthy life expectancy at birth",
    "Freedom to make life choices",
    "Generosity",
    "Positive affect",
    "Confidence in national government",
    "Perceptions of Corruption (Inverse)",
    "Negative Affect (Inverse)",
    "STD Dev of Ladder (Inverse)"
]

# Standardize the data
df_features = df[features]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_features)
```

**K-Means Clustering**

```python
# Determine optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method results
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.xticks(range(1, 11))
plt.grid()
plt.show()

k = 5
kmeans = KMeans(n_clusters=k, random_state=1234)
clusters = kmeans.fit_predict(scaled_data)
df['Comprehensive Wellbeing'] = clusters

# Visualize the clustering
sns.scatterplot(x=df["Log GDP per capita"], y=df["Life Ladder"], hue=df["Comprehensive Wellbeing"], palette="viridis")
plt.title('Clustering of Countries')
plt.xlabel("Log GDP per Capita")
plt.ylabel("Life Ladder")
plt.show()

# Print Cluster Centers and Counts
print("Cluster Centers:\n", kmeans.cluster_centers_)
print("\nCluster Counts:\n", df['Comprehensive Wellbeing'].value_counts())
```

**Feedforward Neural Network**

```python
# Prepare data for neural network
X = scaled_data
y = df['Comprehensive Wellbeing']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Define the model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(k, activation='softmax'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))

# Plot loss and accuracy curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

# Evaluate the model
y_pred = np.argmax(model.predict(X_test), axis=-1)
accuracy = accuracy_score(y_test, y_pred)
precision, recall, fscore, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')

print(f'FNN Accuracy: {accuracy:.4f}')
print(f'FNN Precision: {precision:.4f}')
print(f'FNN Recall: {recall:.4f}')
print(f'FNN F-score: {fscore:.4f}')

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

**Support Network Machine**

```python
# Train SVM model
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)

# Evaluate SVM model
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm, recall_svm, fscore_svm, _ = precision_recall_fscore_support(y_test, y_pred_svm, average='weighted')

print(f'SVM Accuracy: {accuracy_svm:.4f}')
print(f'SVM Precision: {precision_svm:.4f}')
print(f'SVM Recall: {recall_svm:.4f}')
print(f'SVM F-score: {fscore_svm:.4f}')

# Confusion Matrix for SVM
cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (SVM)')
plt.show()
```
